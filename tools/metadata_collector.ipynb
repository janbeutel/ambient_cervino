```python
from dateutil import parser
import pandas as pd

def clean_dates(date_list):
    formatted_dates = []
    for date_str in date_list:
        try:
            if pd.isna(date_str) or str(date_str).strip() == "":
                formatted_dates.append(None)
            else:
                parsed_date = parser.parse(str(date_str))
                formatted_dates.append(parsed_date.strftime('%Y-%m-%d'))
        except Exception:
            formatted_dates.append(None)
    return pd.Series(formatted_dates)

# Apply to your Series (e.g., second_row)
cleaned_dates = clean_dates(second_row)
print(cleaned_dates)
```
